### Criando coluna de "geração"

```python
# Function development to converter intervalls of age to numeric age values.
def age_converter(interval):
    parts = interval.strip('[]').split('-')  
    
    # Checks if the interval is [85+]
    if len(parts) == 1 and parts[0] == '85+':
        return 85
    
    start, end = int(parts[0]), int(parts[1])  
    return (start + end) / 2  

# Function Development to determine the generation with base in the age year.
def determine_generation(year, age):
    # Possibles generations from 1946 to 1964
    
    if 1946 <= year < 1964:
        if age >= 18:
            return 'Silent Generation'
        else:
            return 'Baby Boomers'
    
    # Possibles generations from 1964 to 1980
    
    elif 1964 <= year < 1980:
        if  age >= 40:
            return 'Silent Generation'
        elif age >= 16 and age <= 40:
            return 'Baby Boomers'
        else:
            return 'Generation X'
    
    # Possibles generations from 1980 to 1997
    
    elif 1980 <= year < 1997:
        if   age >= 55:
            return 'Silent Generation'
        elif age >= 34 and age <= 55:
            return 'Baby Boomers'
        elif age >= 17 and age <= 34:
            return 'Generation X'
        else:
            return 'Millennial'
    
    # Possibles generation from 1997 to 2010
    
    elif 1997 <= year < 2010:
        if   age >= 72:
            return 'Silent Generation'
        elif age >= 51 and age <= 72:
            return 'Baby Boomers'
        elif age >= 33 and age <= 51:
            return 'Generation X'
        elif age >= 17 and age <= 33:
            return 'Millennial'
        elif age <= 17:
            return 'Generation Z'
    
    # Possibles generation after 2010
    
    elif year >= 2010: 
        if age >= 85:
            return 'Silent Generation'
        elif  age >= 64 and age <= 85:
            return 'Baby Boomers'
        elif age >= 46 and age <= 64:
            return 'Generation X'
        elif age >= 30 and age <= 46:
            return 'Millennial'
        elif age >= 13 and age <= 30:
            return 'Generation Z'
        else:
            return 'Generation Alpha'
    
# Aplicaning the function in the Dataset "df_suicide_1950_2021_New"
df_suicide_1950_2021_New['Age'] = df_suicide_1950_2021_New['Age Group'].apply(age_converter)
df_suicide_1950_2021_New['Generation'] = df_suicide_1950_2021_New.apply(lambda x: determine_generation(x['Year'], x['Age']), axis=1)
```

### 4.9 Análise Gráfica - Os 10 principais países com maior crescimento na taxa de suicídio desde os anos 2000.

#### Pois bem, nesta última visualização, vou mostrar os países com maior aumento na taxa de suicídio desde a década de 2000, com o objetivo de identificar algum país que esteja apresentando um comportamento incomum em relação aos demais.

```python
# Filter suicide count from the 2000s

suicides_by_country_2000s = df_suicide_1950_2021_New.loc[(df_suicide_1950_2021_New['Year'] >= 2000)]

# Grouping and sum the suicide values by country.

suicides_by_country = suicides_by_country_2000s.groupby('Country Name')['Death rate per 100 000 population'].mean() / 10

# Select the ten countries with large suicide values.
top_10_countries = suicides_by_country.nlargest(10).sort_values()

# Plot bar graphic
plt.figure(figsize=(10,6))
barras = top_10_countries.plot(kind='barh')

for i, v in enumerate(top_10_countries):
    plt.text(v, i, f'{v:.2f}%', ha='left', va='center', fontsize=8, fontweight='bold')

# Add labels to the axis and title. 
plt.xlabel('Número de suicídios por país')
plt.ylabel('País')
plt.title('Os 10 países com maior número de suicídios segundo o número de suicídios por 100.000 habitantes na década de 2000')

plt.xticks(rotation=60)
plt.box(False)
plt.tick_params(axis='x', length=0)
plt.tick_params(axis='y', length=0)

# Show graphic 
plt.show()
```

#### Conforme analisado acima, podemos observar outros países que antes não se destacavam, como a Coreia do Sul, onde mostram que alguns países observaram um aumento nas taxas de suicídio desde a década de 2000.

### 4.9.1 Análise Gráfica - Taxa de Suicídio nos 10 principais países após a década de 2000 (Análise por ano de cada país)

```python
# Filter suicide count from 2000s.
suicides_by_country_2000s = df_suicide_1950_2021_New.loc[df_suicide_1950_2021_New['Year'] >= 2000]

# Grouping and sum the suicide values by country.

suicides_by_country = suicides_by_country_2000s.groupby('Country Name')['Death rate per 100 000 population'].mean() /10

# Select the ten countries with large suicide valores.
top_10_countries = suicides_by_country.nlargest(10).index

# Define the grid layout of subplots

fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(15, 20))
axes = axes.flatten()

# Plot separate graphs for each of the 10 countries

for i, country in enumerate(top_10_countries):
    country_data = suicides_by_country_2000s[suicides_by_country_2000s['Country Name'] == country]
    suicides_by_year = country_data.groupby('Year')['Death rate per 100 000 population'].sum()
    
    axes[i].plot(suicides_by_year.index, suicides_by_year.values, marker='o', linestyle='-')
    axes[i].set_title(f'{country}')
    axes[i].set_xlabel('Ano')
    axes[i].set_ylabel('Taxa de mortalidade por 100.000 habitantes')
    axes[i].tick_params(axis='x', rotation=45)
    axes[i].grid(True)


plt.tight_layout()
plt.show()
```

### Neste gráfico acima, eu mostrei a taxa de suicídio entre os 10 principais países de 2000 a 2021, onde podemos ver o comportamento dos países nesta faixa de anos.
